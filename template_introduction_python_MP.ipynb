## Code Manual

````python
#Deklarasi Data/Variabel
Nama = "Sarah"
NIM = "12345678"
Nilai = 81

print(f"Masukkan Nama Anda: {Nama}")
print(f"Masukkan NIM Anda: {NIM}")
print(f"Masukkan Hasil Ujian (0-100): {Nilai}\n")

#Mengetahui type data berdasarkan variabel
print(f"Nama: {Nama} (type: {type(Nama)})")
print(f"NIM: {NIM} (type: {type(NIM)})")
print(f"Nilai: {Nilai} (type: {type(Nilai)})\n")

#Melakukan Pengkondisian Kategori Sesuai Nilai
if 85 <= Nilai <= 100:
    KategoriNilai = "A (Sangat Baik)"
elif 75 <= Nilai <= 84:
    KategoriNilai = "B (Baik)"
elif 60 <= Nilai <= 74:
   KategoriNilai = "C (Cukup)"
elif 40 <= Nilai <= 59:
    KategoriNilai = "D (Kurang)"
else:
   KategoriNilai = "E (Sangat Kurang)"

#Hasill
print("Hasil Evaluasi Sebagai Berikut:")
print(f"Nama Mahasiswa: {Nama} dengan NIM: {NIM}")
print(f"Nilai Ujian: {Nilai}")
print(f"Kategori Nilai: {KategoriNilai}")
````

## Code dengan `input`

````python
# Input data dari pengguna
nama = input("Masukkan nama mahasiswa: ")
nim = input("Masukkan NIM: ")
nilai = int(input(f"Masukkan nilai ujian (0-100): "))

# Cetak data dengan tipe
print(f"\nNama: {nama} (type: {type(nama)})")
print(f"NIM: {nim} (type: {type(nim)})")
print(f"Nilai: {nilai} (type: {type(nilai)})")

# Menentukan kategori nilai
if 85 <= nilai <= 100:
    kategori = "A (Sangat Baik)"
elif 70 <= nilai < 85:
    kategori = "B (Baik)"
elif 55 <= nilai < 70:
    kategori = "C (Cukup)"
elif 40 <= nilai < 55:
    kategori = "D (Kurang)"
else:
    kategori = "E (Sangat Kurang)"

# Output hasil evaluasi
print("\nHasil Evaluasi:")
print(f"Mahasiswa: {nama} (NIM: {nim})")
print(f"Nilai Ujian: {nilai}")
print(f"Kategori Nilai: {kategori}")
````

### Observasi

*   Menggunakan input() untuk membaca data dari pengguna.
*   Mengonversi nilai ujian dari string ke int agar bisa dievaluasi.
*   Menggunakan if-elif untuk menentukan kategori nilai sesuai rentang.
*   Menampilkan semua informasi termasuk tipe data sesuai permintaan.

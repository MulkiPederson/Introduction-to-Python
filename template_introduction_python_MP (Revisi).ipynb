## Buatlah sebuah program yang:

   1. Meminta input dari pengguna:

       - Nama mahasiswa

       - NIM mahasiswa

       - Nilai ujian (dalam angka, 0â€“100)

   2. Cek dan tampilkan tipe data dari setiap input.

   3. Gunakan if-elif-else untuk menentukan kategori nilai, dengan kriteria sebagai berikut:
   
```
Nilai | Kategori
85 - 100 | A (Sangat Baik)
75 - 84 | B (Baik)
60 - 74 | C (Cukup)
40 - 59 | D (Kurang)
< 40 | E (Sangat Kurang)
```

```
==== Ekspektasi Output Jawaban : =====

Hasil Evaluasi:

Nilai Ujian: 81
Kategori Nilai: B (Baik)
```

## Jawaban (Revisi)

````python
# Input data dari pengguna
nama = input("Masukkan nama mahasiswa: ")
nim = input("Masukkan NIM: ")
nilai = float(input(f"Masukkan nilai ujian (0-100): ")) # Revisi 1. mengganti int ke float karena masih memungkinkan adanya nilai desimal pada inputan nilai

# Cetak data dengan tipe
print(f"\nNama: {nama} (type: {type(nama)})")
print(f"NIM: {nim} (type: {type(nim)})")
print(f"Nilai: {nilai} (type: {type(nilai)})")

# Menentukan kategori nilai
if 85 <= nilai <= 100:
    kategori = "A (Sangat Baik)"
elif 70 <= nilai < 85:
    kategori = "B (Baik)"
elif 55 <= nilai < 70:
    kategori = "C (Cukup)"
elif 40 <= nilai < 55:
    kategori = "D (Kurang)"
elif 0 <= nilai < 40:
    kategori = "E (Sangat Kurang)"      # Revisi 2. Untuk mengatasi inputan nilai diluar dari rentang 0 - 100, dengan menambahkan else diluar dari definisi kategori.
else:
    kategori = ("Nilai Invalid")

# Output hasil evaluasi
print("\nHasil Evaluasi:")
print(f"Mahasiswa: {nama} (NIM: {nim})")
print(f"Nilai Ujian: {nilai}")
print(f"Kategori Nilai: {kategori}")
````

### Observasi

*   Menggunakan input() untuk membaca data dari pengguna.
*   Mengonversi nilai ujian dari string ke float agar bisa dievaluasi, karena masih memungkinkan nilai ujian berupa angka desimal.
*   Menggunakan if-elif untuk menentukan kategori nilai sesuai rentang.
*   Menampilkan semua informasi termasuk tipe data sesuai permintaan.
